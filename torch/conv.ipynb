# 畳み込み層とは

CNN: convolutional neural network（畳込みニューラルネットワーク）

全結合層はflattenにすることにより、配列（形状）を無視した形になる。これってほんとにいいの？
畳み込み層は形状を維持する。3次元のものはそのままにする。
重みはフィルターの中身にあたる。
パディング：画像の周りに値を足す。0が多い？フィルターによって縮小していくと、何度も層に通すことで小さくなってしまう。パッディングの主な目的はこれを無くすこと。  
ストライド：フィルターを移動させる間隔。値が大きいと出力は小さくなる。
チャネルを保つためには、Ｎ個のフィルターが必要らしいのだが、これってどういうことだ？Ｎ個の定義とか書いてあった？
バイアスも同じＮ個らしい。チャネルごとに存在するらしい。  
バッチ処理：i個のまとめたデータに対して同じ？フィルターをかける
プーリング層：空間を小さくする。maxやave。ストライド（スライドの量）はプーリングのsizeと同じ。重ならないように。そうしないバージョンもありそうだけど。最大値の位置が少し変わってもプーリングの範囲に収まっていれば最大値や平均値は変わらない（ロバスト）
